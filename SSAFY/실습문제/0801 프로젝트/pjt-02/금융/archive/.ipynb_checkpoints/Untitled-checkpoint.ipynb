```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Numpy를 활용한 csv file open
def file_open_by_numpy():
    # np.loadtxt(구분자 = ',', 데이터 타입: string)
    np_arr = np.loadtxt('NFLX.CSV', delimiter=",", encoding='cp949', dtype=str)
    return np_arr
```

```python
arr = file_open_by_numpy()
print(arr)
```

```python
# Pandas DataFrame 생성
#기본적인 데이터 프레임 생성
df = pd.DataFrame(arr)
df
```

```python
# 컬럼명 지정하면서 생성하기
columns = arr[0]
arr = np.delete(arr, 0, 0)
df = pd.DataFrame(arr, columns=columns)
```

```python
# Column 이름을 배열 형태로 입력하면 DataFrame 형태(2차원)가 리턴된다.
df = df.loc[:,['Date', 'Open', 'High', 'Low', 'Close']]
df
```

```python
# 2021년 이후 데이터 필터링
# 날짜 열을 인덱스 열로 설정하여 통합 df.loc[start_date:end_date]방법을 사용하여 위의 프로세스를 단순화
# df = df.set_index(['Date'])
# filtered = df.loc['2021-01-01':]
# filtered
```

```python
# 2021년 이후 데이터 필터링
# 시작 날짜와 종료 날짜를 기준으로 boolean mask를 정의하고  df.loc[mask]로 지정
mask = (df['Date'] >= '2021-01-01')
filtered = df.loc[mask]
filtered
```

```python
# 2021년 이후 데이터 필터링
# 시작 날짜와 종료 날짜를 기준으로 boolean mask를 정의하고  df.loc[mask]로 지정
mask = (df['Date'] >= '2021-01-01')
filtered = df.loc[mask]
filtered
```

```python
# 2021년 이후 데이터에서DataFrame의 종가 데이터를 그래프로 시각화
# 종가에 대한 필드를 숫자 형식으로 변환
filtered['Close'] = filtered['Close'].astype(float)

# 종가에 대한 라인 그래프 출력
filtered['Date'] = pd.to_datetime(filtered['Date']) # 년/월 출력 자료형 변경
x = filtered['Date']
y = filtered['Close']
plt.plot(x,y)

# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')
plt.xticks(rotation=45) # 45도 출력 각도 변경

# y축 레이블 설정
plt.ylabel('Close Price')

# 그래프 출력
plt.show()
```

```python
# 최고/최저 종가 추출
max_price = max(filtered['Close'])
min_price = min(filtered['Close'])

print("최고 종가:", max_price)
print("최저 종가:", min_price)
```

```python
# 2022년 이후 데이터를 바탕으로 3가지 지표(월별 최고/최저/종가)를 하나의 그래프에 시각화 

# 2022년 이후 데이터 필터링
mask = (df['Date'] >= '2022-01-01')
filtered = df.loc[mask]
# Open 열 제외
filtered.loc[:, df.columns !='Open']

# 시각화
# 년/월 출력 자료형 변경
filtered['Date'] = pd.to_datetime(filtered['Date']) 

# 종가에 대한 필드를 숫자 형식으로 변환
filtered['High'] = filtered['High'].astype(float)
filtered['Low'] = filtered['Low'].astype(float)
filtered['Close'] = filtered['Close'].astype(float)

# 그래프 만들기
x = filtered['Date'] 
y = filtered['High']
z = filtered['Low']
e = filtered['Close']
plt.plot(x,y, label = 'High')
plt.plot(x,z, label = 'Low')
plt.plot(x,e, label = 'Close')

# 범례 추가
plt.legend()

# 그래프 제목 설정
plt.title('High, Low, and Close Prices January 2022')

# x축 레이블 설정
plt.xlabel('Date')
plt.xticks(rotation=45) # 45도 출력 각도 변경

# y축 레이블 설정
plt.ylabel('Price')

# 그래프 출력
plt.show()

```
